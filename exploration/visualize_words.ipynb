Creates histograms for the word frequencies of the top-n words for a set of documents. 


```python
from exploration.tfidf import computeTFIDF, txtToStrings, topN
```

```python
top_n = topN(*computeTFIDF(txtToStrings()),200)
#print(top_n)
```

### Load the desired 10Qs  and text into a list -> Apple 2018 Q2, AMD Q1, TI 2018 Q1

```python
doc_idcs = [0, 29, 165]
doc_titles = ["Apple 2018 Q2", "AMD 2018 Q1", "TI 2018 Q1"]

all_urls = open('resources/links.txt').read().splitlines()
docs = []
for i, title in zip(doc_idcs, doc_titles):
    fname = "resources/filtered/" + str(i) + ".txt"
    with open(fname,encoding='utf-8') as f:
        print(fname)
        txt_file_as_string = f.read()
    docs.append((txt_file_as_string, title))
```

```python
# imports for counting and plotting histograms 
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
k = 10
mpl.style.use("seaborn")
for doc, doc_title in docs:
    word_freqs = {}
    for top_word in top_n[:k]:
        word_freqs[top_word] = doc.count(top_word)
    labels, values = zip(*word_freqs.items())
    # sort values in descending order
    indSort = np.argsort(values)[::-1]

    # rearrange your data
    labels = np.array(labels)[indSort]
    values = np.array(values)[indSort]
    indexes = np.arange(len(labels))
    bar_width = 0.35
    plt.bar(indexes, values)
    plt.xticks(indexes-.5 + bar_width, labels, rotation='30')
    plt.title("Frequency of Top " + str(k) + " TFIDF Terms in " + doc_title)
    plt.ylabel("Word Frequency")
    plt.show()
    
        
```
